```python
from sklearn import datasets
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mplt
import seaborn as sns
sns.set()
RandomState = 21521109
```

```python
colors = sns.color_palette("Paired").as_hex()
```

```python
X,y = datasets.make_blobs(n_samples=[200,200], n_features=2, random_state=21521109)
kMeans = KMeans(n_clusters=2).fit(X)
centers =  kMeans.cluster_centers_
plt.scatter(X[:,0], X[:,1],c = kMeans.labels_,cmap=mplt.colors.ListedColormap(colors))
plt.scatter(centers[:,0], centers[:,1], marker = 'X', s = 100, color = 'red')
plt.title('Kmeans on blobs dataset')
plt.savefig(fname = 'KmeansExample.png', dpi=500, transparent=False,bbox_inches='tight')
plt.show()
```

```python
Xc, yc = datasets.make_circles(n_samples=400, noise=0.01, random_state=21521109, factor=0.5)
kMeansc = KMeans(n_clusters=2).fit(Xc)
centersc =  kMeansc.cluster_centers_
plt.scatter(Xc[:,0], Xc[:,1],c = kMeansc.labels_ ,cmap=mplt.colors.ListedColormap(colors))
plt.scatter(centersc[:,0], centersc[:,1], marker = 'X', s = 100, color = 'red')
plt.title("Kmeans on circles datasets")
plt.savefig(fname = 'KmeansExampleCircle.png', dpi=500, transparent=False,bbox_inches='tight')

plt.show()
```

```python
def ElbowPlot(data,ncluseter = 2, start = 1, end = 6, Random_state = None, Action = 'show', Name = 'CostPlot.png'):
    inertias = []
    k_candidates = range(start,end+1)
    for k in k_candidates:
        kMeans = KMeans(n_clusters=k).fit(data)
        inertias.append(kMeans.inertia_)
    plt.xlabel('K (no.of clusters)')
    plt.ylabel('J cost function')
    plt.title('Elbow method')
    plt.gca().axes.yaxis.set_ticklabels([])
    sns.scatterplot(x=k_candidates, y=inertias)
    sns.lineplot(x=k_candidates, y=inertias)
    if (Action == 'save'):
        plt.savefig(fname = Name, dpi=500, transparent=False,bbox_inches='tight')
```

```python
ElbowPlot(data=Xc, Action='show')
```
